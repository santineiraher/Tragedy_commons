```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
G= nx.barabasi_albert_graph(10,6, seed = 123)
nx.draw(G, with_labels=True)
```

```python
charact_dict = {'I': (6, 'blue'), 'C': (4, 'green'), 'F': (9, 'red')}
d_central = nx.degree_centrality(G)

dict_nodes = {}
for i in range(G.number_of_nodes()):
    dict_nodes[f'node_{i}'] = {'type': 'I', 'consump': charact_dict['I'][0], 'n_c_neighbors': 0, 'd_centrality': d_central[i], 'color': charact_dict['I'][1]}
```

```python
dict_nodes
```

```python
attrs = {i: dict_nodes[f'node_{i}'] for i in range(G.number_of_nodes())}
nx.set_node_attributes(G, attrs)
```
